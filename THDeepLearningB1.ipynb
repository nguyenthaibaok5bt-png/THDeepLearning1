```python
df1 = pd.DataFrame({'Name' : ['John', 'Smith', 'Paul'], 'Age' : [25, 30, 50]}, index=[0,1,2])
df2 = pd.DataFrame({'Name' : ['Adam', 'Smith'], 'Age' : [26, 11]}, index = [3,4])
df_concat = pd.concat([df1, df2])
print('Concate two DataFrame\n', df_concat)
print('Drop duplicatevalues in DataFrame\n', df_concat.drop_duplicates('Name'))
print('Sort values in DataFrame\n', df_concat.sort_values('Age'))
df_rename = df_concat.rename(columns={'Name': 'Surname', 'Age': 'Age_pupil'})
print('Rename columns of DataFrame\n', df_rename)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/sample_data/23_HOMES.csv')

print("--- 5 dòng đầu tiên của dữ liệu ---")
display(df.head())

print("\n--- Thông tin dữ liệu ---")
print(df.info())

print("\n--- Số lượng giá trị thiếu trong mỗi cột ---")
print(df.isnull().sum())

duplicates = df.duplicated().sum()
print(f"\nSố lượng dòng trùng lặp: {duplicates}")
if duplicates > 0:
    df = df.drop_duplicates()
    print("Đã xóa các dòng trùng lặp.")

print("\n--- Thống kê mô tả ---")
display(df.describe())

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Ma trận tương quan giữa các thuộc tính")
plt.show()

df['Price_per_Area'] = df['Selling_Price'] / df['Area']

print("\n--- Dữ liệu sau khi xử lý sơ bộ ---")
display(df.head())
```
